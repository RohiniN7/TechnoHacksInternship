# Task 2 : Summarizing Data using Pivot Table

# Dependencies

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
train_data = pd.read_csv('train.csv')
```

# Data Frame

```python
train_data
```

# Pivot Table

```python
pivot_table = pd.pivot_table(train_data, index=['OverallQual'])
pd.set_option('display.max_columns', None)
pivot_table
```

```python
pivot_table = pd.pivot_table(train_data, index=['OverallQual'], values=['GrLivArea','SalePrice'])
pivot_table
```

# Bar Chart

## Mean SalePrice for each of the OverallQual

```python
pivot_table = pd.pivot_table(train_data, index=['OverallQual'], values=['SalePrice'])
pivot_table.plot(kind='bar');
```

# Histogram

```python
pivot_table = pd.pivot_table(train_data, index=['OverallQual'], values=['SalePrice'])
pivot_table.plot(kind='hist');
```

# Density Plot

```python
pivot_table = pd.pivot_table(train_data, index=['OverallQual'], values=['SalePrice'])
pivot_table.plot(kind='density');
```

# Area Plot

```python
pivot_table = pd.pivot_table(train_data, index=['OverallQual'], values=['SalePrice'])
pivot_table.plot(kind='area');
```
